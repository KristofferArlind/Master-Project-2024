```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
#NO ADDING TO DATE, RAW

#Daily
fed_funds_rate = pd.read_parquet('../../data/indicators/US/fed_funds_rate.parquet')
fed_funds_rate["date"] = pd.to_datetime(fed_funds_rate["date"])
fed_funds_rate["rate_fed_funds"] = pd.to_numeric(fed_funds_rate["rate_fed_funds"], errors="coerce")
print(fed_funds_rate["date"].min())

fed_yields = pd.read_parquet('../../data/indicators/US/fed_yields.parquet')
fed_yields = fed_yields[["date", "rate_1_year", "rate_3_year", "rate_5_year", "rate_10_year", "rate_20_year"]]
fed_yields["rate_1_year"] = pd.to_numeric(fed_yields["rate_1_year"], errors="coerce")
fed_yields["rate_3_year"] = pd.to_numeric(fed_yields["rate_3_year"], errors="coerce")
fed_yields["rate_5_year"] = pd.to_numeric(fed_yields["rate_5_year"], errors="coerce")
fed_yields["rate_10_year"] = pd.to_numeric(fed_yields["rate_10_year"], errors="coerce")
fed_yields["rate_20_year"] = pd.to_numeric(fed_yields["rate_20_year"], errors="coerce")
fed_yields["date"] = pd.to_datetime(fed_yields["date"])
print(fed_yields["date"].min())

vix = pd.read_parquet('../../data/indicators/US/VIX_daily.parquet')
vix["date"] = pd.to_datetime(vix["date"])
print(vix["date"].value_counts())
print(vix["date"].min())

#Weekly
initial_claims = pd.read_parquet('../../data/indicators/US/initial_claims.parquet')
initial_claims["date"] = pd.to_datetime(initial_claims["date"])
print(initial_claims["date"].value_counts())


#Monthly
credit_spread = pd.read_parquet('../../data/indicators/US/moodys_credit.parquet')
credit_spread["date"] = pd.to_datetime(credit_spread["date"])
credit_spread = credit_spread[["date", "spread"]]
credit_spread.rename(columns={"spread": "credit_spread"}, inplace=True)
print(credit_spread["date"].value_counts())

print(credit_spread["date"].min())

m1m2 = pd.read_parquet('../../data/indicators/US/M1M2.parquet')
m1m2["date"] = pd.to_datetime(m1m2["date"])
print(m1m2["date"].value_counts())
print(m1m2["date"].min())

consumption = pd.read_parquet('../../data/indicators/US/pce.parquet')
consumption["date"] = pd.to_datetime(consumption["date"])
print(consumption["date"].min())
print(consumption["date"].value_counts())


unemployment = pd.read_parquet('../../data/indicators/US/unemployment.parquet')
unemployment["date"] = pd.to_datetime(unemployment["date"])
print(unemployment["date"].min())
print(unemployment["date"].value_counts())

cpi = pd.read_parquet('../../data/indicators/US/cpi.parquet')
cpi["date"] = pd.to_datetime(cpi["date"])
cpi["inflation"] = cpi["cpi"].pct_change(12)*100
print(cpi["date"].min())
print(cpi["date"].value_counts())


ism_index = pd.read_parquet('../../data/indicators/US/ism_index.parquet')
ism_index["date"] = pd.to_datetime(ism_index["date"])
print(ism_index["date"].min())
print(ism_index["date"].value_counts())

dividend_yield = pd.read_parquet('../../data/indicators/US/sp500_dividend_yield.parquet')
dividend_yield["date"] = pd.to_datetime(dividend_yield["date"])
print(dividend_yield["date"].min())
print(dividend_yield["date"].value_counts())

dvps = pd.read_parquet('../../data/indicators/US/sp500_dvps.parquet')
dvps["date"] = pd.to_datetime(dvps["date"])
print(dvps["date"].min())
print(dvps["date"].value_counts())

#Quarterly
real_gdp = pd.read_parquet('../../data/indicators/US/real_gdp.parquet')
real_gdp["date"] = pd.to_datetime(real_gdp["date"])
print(real_gdp["date"].min())
print(real_gdp["date"].value_counts())


real_gnp = pd.read_parquet('../../data/indicators/US/real_gnp.parquet')
real_gnp["date"] = pd.to_datetime(real_gnp["date"])
print(real_gnp["date"].min())
print(real_gnp["date"].value_counts())


#Monthly and Quarterly
earnings_yield = pd.read_parquet('../../data/indicators/US/sp500_earnings_yield.parquet')
earnings_yield["date"] = pd.to_datetime(earnings_yield["date"])
print(earnings_yield["date"].min())
print(earnings_yield["date"].value_counts())

eps = pd.read_parquet('../../data/indicators/US/sp500_eps.parquet')
eps["date"] = pd.to_datetime(eps["date"])
print(eps["date"].min())
print(eps["date"].value_counts())
```

```python
#Merge all:
all_indicators = fed_funds_rate.copy()
all_indicators = all_indicators.merge(fed_yields, on="date", how="outer")
all_indicators = all_indicators.merge(vix, on="date", how="outer")
all_indicators = all_indicators.merge(initial_claims, on="date", how="outer")
all_indicators = all_indicators.merge(credit_spread, on="date", how="outer")
all_indicators = all_indicators.merge(m1m2, on="date", how="outer")
all_indicators = all_indicators.merge(consumption, on="date", how="outer")
all_indicators = all_indicators.merge(unemployment, on="date", how="outer")
all_indicators = all_indicators.merge(cpi, on="date", how="outer")
all_indicators = all_indicators.merge(ism_index, on="date", how="outer")
all_indicators = all_indicators.merge(dividend_yield, on="date", how="outer")
all_indicators = all_indicators.merge(dvps, on="date", how="outer")
all_indicators = all_indicators.merge(real_gdp, on="date", how="outer")
all_indicators = all_indicators.merge(real_gnp, on="date", how="outer")
all_indicators = all_indicators.merge(earnings_yield, on="date", how="outer")
all_indicators = all_indicators.merge(eps, on="date", how="outer")

```

```python
all_indicators.to_parquet('../../data/indicators/US/all_indicators_raw_outer.parquet')
```
